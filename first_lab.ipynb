# **1.1 Вероятностное пространство, формула Байеса.**

# **№ 1.**

(a) Несовместные события.  

Пусть $A \cap B = \varnothing$ (несовместны) и $P(A) > 0$, $P(B) > 0$, то есть события возможны. Тогда $P(A \cap B) = 0$, но $P(A)P(B) > 0$. Следовательно, $P(A \cap B) \neq P(A)P(B)$ -- события зависимы.

(b) События, образующие $\sigma$-алгебру $\Sigma$.  

Образование $\sigma$-алгебры $\Sigma$ накладывает условия на замкнутость относительно операций, но не на вероятностные связи, то есть, события могут бять и зависимы, и независимы. В $\Sigma$ всегда содержатся $\varnothing$ и $\Omega$, и $P(\varnothing \cap \Omega) = 0 \neq P(\varnothing)P(\Omega) = 0$.

(c) События, имеющие одинаковую вероятность.  

То, что $P(A) = P(B)$, не является ни необходимым, ни достаточным условием независимости. Если $A = B$ и $0 < P(A) < 1$, то $P(A) = P(B)$, но $P(A \cap B) = P(A) \neq P(A)P(B)$, и отсюда вывод, что события зависимы.

# **№ 2.**

(a) $A + C = E$  
$A = \{ГГ, ГР\}$ (герб на 1й монете), $C = \{ГГ, РГ\}$ (герб на 2й монете).  
Объединение: $\{ГГ, ГР, РГ\} = E$ (хотя бы один герб).

(b) $AC = K$  
Пересечение $A$ и $C$: $\{ГГ, ГР\} \cap \{ГГ, РГ\} = \{ГГ\} = K$ (два герба).

(c) $EF = G$  
$E = \{ГГ, ГР, РГ\}$ (хотя б один герб), $F = \{ГР, РГ, РР\}$ (хотя б одна решка).  
Пересечение: $\{ГР, РГ\} = G$ (один герб и одна решка).

(d) $G + E = E$  
$G = \{ГР, РГ\} \subset E = \{ГГ, ГР, РГ\}$, поэтому объединение равно $E$.

(e) $GE = G$  
$G \cap E = \{ГР, РГ\} \cap \{ГГ, ГР, РГ\} = \{ГР, РГ\} = G$.

(f) $BD = H$  
$B = \{РГ, РР\}$ (решка на 1-й), $D = \{ГР, РР\}$ (решка на 2й).  
Пересечение: $\{РР\} = H$ (ни одного герба).

(g) $E + K = E$  
$K = \{ГГ\} \subset E = \{ГГ, ГР, РГ\}$, поэтому объединение $= E$.

# **№ 5.**

1. при $m = 1$:    

Одного снаряда хватит для поражения самолета только в случае попадания в кабину и двигатель. Получается,
$$\mathsf{P}(A|m=1) = p_1.$$

2. при $m = 2$:    

Двух снарядов хватит для поражения, если хотя бы один попал в первую часть $A_1$. Вероятность обратного события (то есть в первую часть не попали):    
$$P(\bar{A}_1|m = 2) = (p_2 + p_3)^2.$$

 Также самолет уничтожат два попадания во вторую часть $A_2$.

Самолет выстоит, если из вероятности попадания двух снарядов во вторую часть вычесть вероятность попадания в кабину. Получается, что
$$P(A|m = 2) = 1 - (2p_2p_3 + p_3^2).$$

3. при $m = 3$:    

Самолет не будет поражен, если один из снарядов попадет в третью часть, а другие два - во вторую. Получаем, что

$$\mathsf{P}(A|m=3) = 1-3\mathsf{p_2}{p_3}^2.$$

4. при $m = 4$:    

Самолет в любом случае уничтожат.
$$\mathsf{P}(A|m=4) = 1.$$


# **1.2 Случайный вектор и числовые характеристики.**

# **№ 1.**

$f_\xi(x, y) = \frac{e^{-2|y|}}{\pi(1+x^2)}$

Для проверки, является ли заданная функция плотностью распределения случайного вектора, нужно проверить условия неотрицательнотси и нормированности.

1. Неотрицательность.   

Функция представляет собой произведение:

$\frac{1}{\pi(1+x^2)} > 0$ для всех $x \in \mathbb{R}$

$e^{-2|y|} > 0$ для всех $y \in \mathbb{R}$

Следовательно, $f_\xi(x, y) > 0$ для всех $(x, y) \in \mathbb{R}^2$

2. Нормированность.    

$$
\iint_{\mathbb{R}^2} f_\xi(x, y) dxdy = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \frac{e^{-2|y|}}{\pi(1+x^2)} dxdy
$$

функция распадается на произведение функций от $x$ и от $y$:
$$
= \frac{1}{\pi} \left(\int_{-\infty}^{\infty} \frac{1}{1+x^2} dx\right) \left(\int_{-\infty}^{\infty} e^{-2|y|} dy\right)
$$

первый интеграл:
$$
\int_{-\infty}^{\infty} \frac{1}{1+x^2} dx = \arctan x \Big|_{-\infty}^{\infty} = \pi
$$

второй интеграл:
$$
\int_{-\infty}^{\infty} e^{-2|y|} dy = 2\int_{0}^{\infty} e^{-2y} dy = 1
$$

итого:
$$
\iint_{\mathbb{R}^2} f_\xi(x, y) dxdy = \frac{1}{\pi} \cdot \pi \cdot 1 = 1
$$

Получилось, что функция является плотностью распределения случайного вектора.

# **№ 2.**   
Дано совместное распределение случайных величин $\xi$ и $\eta$:

$$
\begin{array}{c|ccc}
\eta \setminus \xi & -1 & 0 & 1 \\ \hline
-1 & \frac{1}{8} & \frac{1}{12} & \frac{7}{24} \\
1 & \frac{1}{3} & \frac{1}{6} & 0
\end{array}
$$

(a) Маргинальные распределения

Суммируем вероятности по строкам и столбцам:

Распределение $\xi$:
$$
\begin{array}{c|ccc}
\xi & -1 & 0 & 1 \\ \hline
P & \frac{11}{24} & \frac{1}{4} & \frac{7}{24}
\end{array}
$$

Распределение $\eta$:
$$
\begin{array}{c|cc}
\eta & -1 & 1 \\ \hline
P & \frac{1}{2} & \frac{1}{2}
\end{array}
$$

(b) Матожидание и матрицы

Матожидание:
$$
E\xi = -1\cdot\frac{11}{24} + 0\cdot\frac{1}{4} + 1\cdot\frac{7}{24} = -\frac{1}{6}
$$
$$
E\eta = -1\cdot\frac{1}{2} + 1\cdot\frac{1}{2} = 0
$$

Дисперсия и ковариация:
$$
D\xi = E\xi^2 - (E\xi)^2 = \frac{3}{4} - \frac{1}{36} = \frac{13}{18}
$$
$$
D\eta = E\eta^2 - (E\eta)^2 = 1 - 0 = 1
$$
$$
cov(\xi,\eta) = E(\xi\eta) - E\xi E\eta = -\frac{1}{2} - (-\frac{1}{6})\cdot 0 = -\frac{1}{2}
$$

Ковариационная матрица:
$$
\begin{pmatrix}
D\xi & cov(\xi,\eta) \\
cov(\xi,\eta) & D\eta
\end{pmatrix} =
\begin{pmatrix}
\frac{13}{18} & -\frac{1}{2} \\
-\frac{1}{2} & 1
\end{pmatrix}
$$

Корреляционная матрица:
$$
\rho(\xi,\eta) = \frac{cov(\xi,\eta)}{\sqrt{D\xi}\sqrt{D\eta}} = \frac{-\frac{1}{2}}{\sqrt{\frac{13}{18}}\cdot 1} \approx -0.588
$$
$$
\begin{pmatrix}
1 & \rho(\xi,\eta) \\
\rho(\xi,\eta) & 1
\end{pmatrix} =
\begin{pmatrix}
1 & -0.588 \\
-0.588 & 1
\end{pmatrix}
$$

(c) Независимость и некоррелированность

Т.к. $\rho(\xi,\eta) \neq 0$, величины коррелированы.

Проверка независимоси: $P(\xi=-1,\eta=-1) = \frac{1}{8}$,
$P(\xi=-1)\cdot P(\eta=-1) = \frac{11}{24}\cdot\frac{1}{2} = \frac{11}{48}$

Т.к. $\frac{1}{8} \neq \frac{11}{48}$, случайные величины зависимы.
